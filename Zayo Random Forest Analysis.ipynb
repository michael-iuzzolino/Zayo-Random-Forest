# Data Munging

## Accounts

### Reading in Data and Creating Dataframe

```python
# Read in the CPQs csv
accounts_df = pd.read_csv("data/accounts.csv", sep=",")

# Clean column headers from ufeff
cleaned_accounts_column = []
for acct_col in accounts_df.columns:
    new_head = acct_col.replace(u'\ufeff', '').strip()
    cleaned_accounts_column.append(new_head)

# Assign cleaned column heads to dataframe columns
accounts_df.columns = cleaned_accounts_column

# Delare buildings header
accounts_header = accounts_df.columns
```

### Get counts of missing data
http://pandas.pydata.org/pandas-docs/stable/missing_data.html

```python
# Check for null data (http://pandas.pydata.org/pandas-docs/stable/missing_data.html)

accounts_missing_data_dict = {}
for header in accounts_header:
    accounts_missing_data_dict[header] = 0
    current_column_data = accounts_df[header]
    column_data_isnull = pd.isnull(current_column_data)
    for column_val in column_data_isnull:
        if column_val:
            accounts_missing_data_dict[header] += 1

# Print count of missing data
print("Buildings: Missing Data")
print("{:s}\n".format("="*len("Buildings: Missing Data")))
print("{:30s} {:13s}".format("Feature", "Missing Count"))
print("{:44s}".format("-"*44))
for key, val in accounts_missing_data_dict.items():
    print("{:30s} {:5d}".format(key, val))
```

# Merge!

Merge by AcctID key
http://stackoverflow.com/questions/23560789/merging-two-data-frames-by-key-column


https://chrisalbon.com/python/pandas_join_merge_dataframe.html

```python
def drop_x(df, drop_end="_x"):
    # list comprehension of the cols that end with '_y'
    to_drop = [x for x in df if x.endswith(drop_end)]
    df.drop(to_drop, axis=1, inplace=True)
```

```python
# Opportunities and Building df merger on key "Building ID"
opp_bldg_df = pd.merge(opportunities_df, buildings_df, on="Building ID")

# Merge opp_build with accounts by "Account ID"
acct_opp_bldg = pd.merge(opp_bldg_df, accounts_df, on="Account ID")

# Merge acct_opp_bldg with cpqs
acct_cpq_opp_bldg = pd.merge(acct_opp_bldg, cpqs_df, on="Account ID")

# # clean header
# cleaned_header = []
# for head in acct_cpq_opp_bldg.columns:
#     cleaned_header.append(head.replace("_x", "").replace("_y", ""))

# acct_cpq_opp_bldg.columns = cleaned_header

drop_x(acct_cpq_opp_bldg, '_x')
drop_x(acct_cpq_opp_bldg, '_y')

acct_cpq_opp_bldg.columns
```

```python
def sort_pd(key=None,reverse=False,cmp=None):
    def sorter(series):
        series_list = list(series)
        return [series_list.index(i) 
           for i in sorted(series_list,key=key,reverse=reverse,cmp=cmp)]
    return sorter
```

# Set ML Training and Testing Data

## X and Y
The labels to compare y_hat to are in column "StageName"

```python
# X
term_in_months = df_filtered_6["Term in Months"]
opportunity_type = df_filtered_6["Opportunity Type"]
net_class = df_filtered_6["Net Classification"]
type_business = df_filtered_6["Type"]
estimated_build_cost = df_filtered_6["Estimated Build Cost"]
x36_mrc = df_filtered_6["X36 MRC List"]
x36_nrr = df_filtered_6["X36 NRR List"]
x36_npv = df_filtered_6["X36 NPV List"]
market_ = df_filtered_6["Market"]
city_ = df_filtered_6["City"]
postal_code = df_filtered_6["Postal Code"]
net_prox = df_filtered_6["Network Proximity"]
on_zayo_stat = df_filtered_6["On Zayo Network Status"]
industry = df_filtered_6["Industry"]
vertical = df_filtered_6["Vertical"]
total_brr = df_filtered_6["Total BRR"]
annual_rev = df_filtered_6["AnnualRevenue"]
num_employ = df_filtered_6["NumberOfEmployees"]
dnb_rev = df_filtered_6["DandB Revenue"]
dnb_tot_emp = df_filtered_6["DandB Total Employees"]

              

# Y
stage_name = df_filtered_6["StageName"]

# Determines where y is sorted
converter = {"Closed - Lost" : 0, 
             "1 - Working" : -1, 
             "2 - Best Case" : -1, 
             "3 - Committed" : 1, 
             "4 - Closed" : 1, 
             "5 - Accepted" : 1}

opp_type_dict = { key : index for index, key in enumerate(opportunity_type.unique())}
net_class_dict = { key : index for index, key in enumerate(net_class.unique())}
type_business_dict = { key : index for index, key in enumerate(type_business.unique())}
market_dict = { key : index for index, key in enumerate(market_.unique())}
city_dict = { key : index for index, key in enumerate(city_.unique())}
postal_code_dict = { key : index for index, key in enumerate(postal_code.unique())}
on_zayo_stat_dict = { key : index for index, key in enumerate(on_zayo_stat.unique())}

industry_dict = { key : index for index, key in enumerate(industry.unique())}
vertical_dict = { key : index for index, key in enumerate(vertical.unique())}
```

```python

training_data_x = []
training_data_y = []

test_data_x = []
test_data_id = []

for index, key in enumerate(stage_name):
    classifier = converter[key]
  
    # X
    term = float(term_in_months[index])
    opp = int(opp_type_dict[opportunity_type[index]])
    net = int(net_class_dict[net_class[index]])
    busin_type = int(type_business_dict[type_business[index]])
    est_cost = float(estimated_build_cost[index].strip().replace("$", "").replace(",", ""))
    x36_mrc_temp = float(x36_mrc[index].strip().replace("$", "").replace(",", ""))
    x36_nrr_temp = float(x36_nrr[index].strip().replace("$", "").replace(",", ""))
    x36_npv_temp = float(x36_npv[index].strip().replace("$", "").replace(",", ""))
    market_temp = int(market_dict[market_[index]])
    city_temp = int(city_dict[city_[index]])
    postal_temp = int(postal_code_dict[postal_code[index]])
    net_prox_temp = float(net_prox[index])
    on_zayo_temp = int(on_zayo_stat_dict[on_zayo_stat[index]])
    
    industry_temp = int(industry_dict[industry[index]])
    vertical_temp = int(vertical_dict[vertical[index]])
    total_brr_temp = float(total_brr[index].strip().replace("$", "").replace(",", ""))
    annual_rev_temp = float(annual_rev[index].strip().replace("$", "").replace(",", ""))
    num_employ_temp = int(num_employ[index])
    dnb_rev_temp = float(dnb_rev[index].strip().replace("$", "").replace(",", ""))
    dnb_tot_emp_temp = int(dnb_tot_emp[index])

    
    temp_x = [term, opp, net, busin_type, est_cost, 
              x36_mrc_temp, x36_nrr_temp, x36_npv_temp,
              market_temp, city_temp, postal_temp, net_prox_temp, on_zayo_temp,
              industry_temp, vertical_temp, total_brr_temp, annual_rev_temp, num_employ_temp,
              dnb_rev_temp, dnb_tot_emp_temp]

    # X "test"
    if classifier == -1:
        test_data_x.append(temp_x)
        test_data_id.append(df_filtered_6["Opportunity ID"][index])
    
    # Else
    else:
        # Training X
        training_data_x.append(temp_x)
        # Training Y
        training_data_y.append(classifier)
        
        
```

```python
features = ["term", "opportunity", "net classification", "type", 
            "est cost", "x36 mrc", "x36 nrr", "x36 npv",
            "Market", "City", "Postal Code", "Net Proximity", "On Zayo",
            "Industry", "Vertical", "Total BRR", "AnnualRevenue", "NumberOfEmployees",
            "DandB Revenue", "DandB Total Employees"
           ]


train_x_df = pd.DataFrame(training_data_x, columns=features)
train_y_df = pd.DataFrame(training_data_y, columns=["Sold"])

new_df = pd.concat([train_x_df, train_y_df], axis=1)

# http://stackoverflow.com/questions/18689823/pandas-dataframe-replace-nan-values-with-average-of-columns
train_x_df = new_df.fillna(value=new_df.mean())

test_data_x_df = pd.DataFrame(test_data_x, columns=features)
test_x_df = test_data_x_df.fillna(value=0)

test_data_id_df = pd.DataFrame(test_data_id, columns=["Opportunity ID"])
```

## Clean cost values
- Estimated Build Cost
- Total BRR
- AnnualRevenue
- DandB Revenue
- X36 MRC List 
- X36 NRR List
- X36 NPV List

```python
clf = ensemble.RandomForestClassifier()
fraction = 9.0/10.0
shuffled = train_x_df.sample(frac=1)



features_dict = { val:key for val, key in enumerate(features)}

prediction = ["Sold"]
trainX = shuffled[features][:int(round(fraction*len(shuffled)))].values
trainY = shuffled[prediction][:int(round(fraction*len(shuffled)))].values

trainY = np.reshape(trainY, (len(trainY),))

clf = clf.fit(trainX, trainY)

testX = shuffled[features][int(round(fraction*len(shuffled))):].values
testY = shuffled[prediction][int(round(fraction*len(shuffled))):].values

testY = np.reshape(testY, (len(testY),))
```

```python
print(clf.score(testX, testY))
```

```python
# http://scikit-learn.org/stable/auto_examples/ensemble/plot_forest_importances.html
importances = clf.feature_importances_
std = np.std([tree.feature_importances_ for tree in clf.estimators_],
             axis=0)
indices = np.argsort(importances)[::-1]

# Print the feature ranking
print("Feature ranking:")

for f in range(trainX.shape[1]):
    print("%d. feature %d (%f)" % (f + 1, indices[f], importances[indices[f]]))

    
# Plot the feature importances of the forest
x_labels = [features_dict[val] for val in indices]


plt.figure()
plt.title("Feature importances")
plt.bar(range(trainX.shape[1]), importances[indices],
       color="r", yerr=std[indices], align="center")
plt.xticks(range(trainX.shape[1]), x_labels, rotation='vertical')
plt.xlim([-1, trainX.shape[1]])
plt.show()
```

```python
predict_y = clf.predict(test_x_df)

predict_y_df = pd.DataFrame(predict_y, columns=["Prediction"])


predict_df = pd.concat([test_data_id_df, test_x_df, predict_y_df], axis=1)

good_ops = predict_df[predict_df['Prediction'] == 1]
bad_ops = predict_df[predict_df['Prediction'] == 0]

good_ops_ids = good_ops["Opportunity ID"]
good_ops_ids_unique = good_ops_ids.unique()

bad_ops_ids = bad_ops["Opportunity ID"]
bad_ops_ids_unique = bad_ops_ids.unique()

num_good_ops = len(good_ops_ids_unique)
num_bad_ops = len(bad_ops_ids_unique)
print("Number of good opportunities: {}".format(num_good_ops))
print("Number of bad opportunities: {}".format(num_bad_ops))
```
